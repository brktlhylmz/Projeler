```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("survey lung cancer.csv")
df.head()
```

```python
df["GENDER"].replace(["F","M"],[1,2],inplace=True)
df["LUNG_CANCER"].replace(["YES","NO"],[2,1],inplace=True)
```

```python
df.head()
```

```python
sns.heatmap(df.pivot_table("LUNG_CANCER",index="SMOKING",columns="GENDER"),annot=True,cmap="Blues",fmt=".4")
df.pivot_table("LUNG_CANCER",index="SMOKING",columns="GENDER")
```

```python
plt.figure(figsize=(15,12))
sns.heatmap(df.drop(["GENDER"],axis=1).corr(),annot=True,cmap="Spectral")
```

```python
plt.figure(figsize=(20,10))
sns.swarmplot(data=df, x="AGE",y="LUNG_CANCER")
```
